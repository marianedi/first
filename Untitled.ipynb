```python
from __future__ import print_function
import os
data_path = ['data']
```

```python
import pandas as pd

# Import the data using the file path
filepath = os.sep.join(data_path + ["fraud_detection_bank_dataset.csv" ])
data = pd.read_csv('fraud_detection_bank_dataset.csv')
data
```

```python
data.head(1).T
```

```python
# Remove extraneous columns
data.drop(["col_111","col_1"], axis=1, inplace=True)
data.fillna(0, inplace=True)
```

```python
# Get a list of all the columns that don't contain the label
x_cols = [x for x in data.columns if x != 'targets']

# Split the data into two dataframes
X_data = data[x_cols]
y_data = data['targets']

# # alternatively:
# X_data = data.copy()
# y_data = X_data.pop('churned')
X_data
```

```python
# Get a list of all the columns that don't contain the label
x_cols = [x for x in data.columns if x != 'targets']

# Split the data into two dataframes
X_data = data[x_cols]
y_data = data['targets']

# # alternatively:
# X_data = data.copy()
# y_data = X_data.pop('churned')
y_data 
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn = knn.fit(X_data, y_data)

y_pred = knn.predict(X_data) 
y_pred
```

```python
 def accuracy(real, predict):
    return sum(y_data == y_pred) / float(real.shape[0])
```

```python
print(accuracy(y_data, y_pred))
```

```python
knn = KNeighborsClassifier(n_neighbors=5, weights='uniform')

knn = knn.fit(X_data, y_data)

y_pred = knn.predict(X_data)

print(accuracy(y_data, y_pred))

```

```python
score_list = list()

for k in range(1, 21):
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn = knn.fit(X_data, y_data)
    
    y_pred = knn.predict(X_data)
    score = accuracy(y_data, y_pred)
    
    score_list.append((k, score))
    
score_df = pd.DataFrame(score_list, columns=['k', 'accuracy'])
score_df
```
